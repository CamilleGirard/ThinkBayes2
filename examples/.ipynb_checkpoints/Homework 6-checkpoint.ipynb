### Grid algorithm

```python
class ChangeClass(Suite, Joint):
    
    def Likelihood(self, data, hypo):
        """
        
        data: array of counts
        hypo: t, m1, m2, st1, st2
        """
        t, m1, m2, st1, st2 = hypo
        before = data[:t]
        after = data[t:]
        like1 = np.prod(norm.pdf(before, m1))
        like2 = np.prod(norm.pdf(after,m2))
        return like1*like2
```

```python
ts = range(len(nile))
mus = np.linspace(950,1000,51);
sts = np.linspace(0,10,11);

from itertools import product

hypos = product(ts, mus, mus, sts,sts)
```

```python
classchange = ChangeClass(hypos)
```

```python
classchange.Update(data)
```

```python
tau = classchange.Marginal(0)
thinkplot.Pdf(tau)
```

```python
K = 50
N =Fire.shape[0]

def stick_breaking(beta):
    portion_remaining = T.concatenate([[1], T.extra_ops.cumprod(1 - beta)[:-1]])

    return beta * portion_remaining
```
