## The Nile Problem

The nile rivers trends seem to have quickly changed at some point - this is an analysis of where and how much that change was.

First use a changepoint algorithim in pymc in order to identify when the mean changed and to have the normal distributions for the means in each half. 

```python
import statsmodels.api as sm

data = sm.datasets.nile.load()
```

```python
nile = sm.datasets.get_rdataset("Nile")
nile.data.head()
```

```python
data = nile.data['value']
print(data.head())
```

```python
nile.data['value'].plot()
```

### Mean Based Changepoint

```python
niter = 100000
t = range(len(data))
with pm.Model() as change_point:
    cp = pm.DiscreteUniform('change_point', lower=0, upper=len(data))
    mu0 = pm.Uniform('mu0', 500,1200)
    mu1 = pm.Uniform('mu1', 500,1200)
    mu = T.switch(t < cp, mu0, mu1)
    Y_obs = pm.Normal('Y_obs', mu=mu, sd = data.mean(), observed=data)
    trace = pm.sample(niter)
```

```python
pm.traceplot(trace, varnames=['change_point', 'mu0', 'mu1'])
pass
```

With all the noise the model (caused by the fluctuation of the tide) is not very convinced that the changepoint is at 30, however there is a  small spike. The means are also relatively different, with the earlier mean being higher before the changepoint, fitting with the plot given by the data. 

### Change in the Regression

It also seems that there is a change in the regression before and after the changepoint. For this I will calculate a regression before at after 30, which is around where the change happens, and will see what this shows.

```python
first = nile.data['value'][0:30]
t1 = nile.data['time'][0:30]
second = nile.data['value'][30:]
t2 = nile.data['time'][30:]
```

```python
with pm.Model() as model:
    slope = pm.Uniform('slope', -10, 10)
    inter = pm.Uniform('inter', 100, 10000)
    sigma = pm.Uniform('sigma', 0, 1000)

    y_est = slope* (t1-1870) + inter
    y = pm.Normal('y', mu=y_est, sd=sigma, observed= first)
    trace = pm.sample(4000, tune=20000)
```

```python
pm.traceplot(trace, varnames=['slope', 'inter', 'sigma'])
pass
```

```python
with pm.Model() as model:
    slope = pm.Uniform('slope', -10, 10)
    inter = pm.Uniform('inter', 100, 10000)
    sigma = pm.Uniform('sigma', 0, 1000)

    y_est = slope* (t2 - nile.data['time'][30]) + inter
    y = pm.Normal('y', mu=y_est, sd=sigma, observed=second)
    trace = pm.sample(4000, tune=25000)
```

```python
pm.traceplot(trace, varnames=['slope', 'inter', 'sigma'])
pass
```

With a dividing time of 30, which is somewhere in the range of the bump of the changepoint there is an observed difference in the slope, intercept and the sigma. Before the changepoint there is a negative slope of about -2.5 and a higher intercept of about 1100 and a higher sigma. Afterwards the slope is positive but about 0.5 and the intercept is lower as well as the sigma. 

This fits with the plot of the data, as the beginning the level is dropping starting at a higher point and having a bigger fluctuation. After the point the level has less fluctuation and is more constant. 

### Combining the Regression and Change Point

Instead of doing changepoint with the mean, use two regression parameters to find the changepoint and the components of each side of the changepoint. 

```python
t = range(len(data))
with pm.Model() as model:
    cp = pm.DiscreteUniform('change_point', lower=0, upper=len(data))
    
    s1 = pm.Uniform('slope1', -15, 10)
    i1 = pm.Uniform('inter1', 850, 1400)
    si1 = pm.Uniform('sigma1', 100, 250)
    s2= pm.Uniform('slope2', -4, 4)
    i2=pm.Uniform('inter2', 400, 1200)
    si2 = pm.Uniform('sigma2', 80, 200)
    
    slope = T.switch(t < cp, s1, s2)
    inter = T.switch(t < cp, i1, i2)
    sigma = T.switch(t < cp, si1, si2)
    
    y_est = slope* (nile.data['time']-1870) + inter
    y = pm.Normal('y', mu=y_est, sd=sigma, observed=nile.data['value'])
    trace = pm.sample(100000, tune=100000)
```

```python
pm.traceplot(trace, varnames=['change_point', 'slope1', 'inter1', 'sigma1', 'slope2', 'inter2', 'sigma2'])
pass
```

By combining the changepoint and the regression model there is a clearer changepoint location at 28 than the previous model, due to the reducation of the effect of the fluctuation which is now put into the sigma. The slopes, intercepts and sigmas are very similar to those collected by running the regression on each side of the changepoint of the mu which shows consistancy between the two methods. One larger difference is that the slope 1 has some more fluctuation, it is less sure than when the data is divided, but still has a slope that is leaning into the slightly negative side. 

### We Have 2 Possible Changepoints...

The slope at the beginning of the changepoint being bimodal means that there are two potential changepoints. This is also shown in the changepoint data, where there is a small bump up later in the data for another possible changepoint. In order to observe how these two sides of either changepoint behave we will run the model with both halves.

Lets split the data at 40 for the beginning and start at 35 for the end and run the model

```python
first_half = nile.data['value'][0:41]
time1 = nile.data['time'][0:41]
second_half = nile.data['value'][35:]
time2 = nile.data['time'][35:]
```

```python
t = range(len(first_half))
with pm.Model() as model:
    cp = pm.DiscreteUniform('change_point', lower=0, upper=len(first_half))
    
    s1 = pm.Uniform('slope1', -15, 10)
    i1 = pm.Uniform('inter1', 850, 1400)
    si1 = pm.Uniform('sigma1', 100, 250)
    s2= pm.Uniform('slope2', -10, 25)
    i2=pm.Uniform('inter2', -100, 1200)
    si2 = pm.Uniform('sigma2', 30, 200)
    
    slope = T.switch(t < cp, s1, s2)
    inter = T.switch(t < cp, i1, i2)
    sigma = T.switch(t < cp, si1, si2)
    
    y_est = slope * (time1 - 1870) + inter
    y = pm.Normal('y', mu=y_est, sd=sigma, observed=first_half)
    trace = pm.sample(10000, tune=10000)
```

```python
pm.traceplot(trace, varnames=['change_point', 'slope1', 'inter1', 'sigma1', 'slope2', 'inter2', 'sigma2'])
pass
```

With the changepoint analysis on the first points of the data until 40 the changepoint occurs at 28, where it was said to in the whole analysis. The slope for the beginning of the data is positive, which is not what was expected to happen here, it was expected to be negative. The slope afterwards becomes very positive which is also a little shocking, as it is at about 12. 

Not having a lot of data on either side can contribute to these suprising findings, the changepoint seems accurate with the data, however the regressions do not seems to fit as expected.

## Crime Data Changepoint Problem

This is from the notes on change-point in class. Observing if and when a change happened in the time between homocides and shootings with a firearm.

```python
!wget https://raw.githubusercontent.com/baltimore-sun-data/2018-shootings-analysis/master/BPD_Part_1_Victim_Based_Crime_Data.csv;
```

```python
df = pd.read_csv('BPD_Part_1_Victim_Based_Crime_Data.csv', parse_dates=[0])
df.head()
```

```python
shootings = df[df.Description.isin(['HOMICIDE', 'SHOOTING']) & (df.Weapon == 'FIREARM')]
```

```python
grouped = shootings.groupby('CrimeDate')
```

```python
counts = grouped['Total Incidents'].sum()
counts.head()
```

```python
index = pd.date_range(counts.index[0], counts.index[-1])
counts = counts.reindex(index, fill_value=0)
counts.head()
```

```python
counts.plot()
```

```python
niter = 60000
t = range(len(counts))
with pm.Model() as change_point:
    cp = pm.DiscreteUniform('change_point', lower=0, upper=len(counts), testval=len(counts)//2)
    mu0 = pm.Exponential('lam0', 1/counts.mean())
    mu1 = pm.Exponential('lam1', 1/counts.mean())
    mu = T.switch(t < cp, mu0, mu1)
    Y_obs = pm.Poisson('Y_obs', mu=mu, observed=counts)
    trace = pm.sample(niter)
```

```python
pm.traceplot(trace, varnames=['change_point', 'lam0', 'lam1'])
pass
```

There seems to have been a change in the time between firearm homicides or shootings after the changepoint. The changpoint occurs between day 1200 and 1215, afterwards the time between firearm crimes seems to go up to about 2.7 from about 1.5. The lamdas have little difference between each sample however the change point seems to not be at a specific date but contained between a set of about 15 days.

16 years represents a native speaker, this is just an assumption about the amount of years that the student has been able to speak the language.
