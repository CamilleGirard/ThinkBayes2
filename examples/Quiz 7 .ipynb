modeled by a two-parameter gamma distribution. There are several ways to parameterize the
gamma distribution; we’ll use the one with a shape parameter, k, and a scale parameter, θ,

where Γ is the gamma function.

```python
x = 2
k = 3
t = 2
gamma = scipy.special.gamma(k)
pdf = 1/((gamma)*t**k)* x**(k-1)* np.exp(-x/t)
print(pdf)
```

2. During the last three days in the Boston area, we have measured the following rainfalls in
inches: 0.78, 0.87, 0.64.
Use this data to compute a joint posterior distribution for the parameters of the gamma
distribution.
You can use the following priors:
• For k, Half-normal with σ = 0.5.
• For θ, Half-normal with σ = 4.
What are the posterior means for k and θ?

```python
data = [0.78, 0.87, 0.64]
```

```python
with pm.Model() as model:
    k = pm.distributions.continuous.HalfNormal('k',0.5)
    t = pm.distributions.continuous.HalfNormal('t',4)
    
    pdf = pm.InverseGamma('x', alpha = k, beta = t, observed= data)
    trace = pm.sample(4000, tune=2000)
```

```python
pm.traceplot(trace);
```

```python
summary = pm.summary(trace)
print(summary)
```

```python
ax = pm.plot_posterior(trace)
```

3. Generate a predictive distribution for the amount of rain we will get tomorrow (assuming
that it rains at all). What is the predictive mean?

```python
prior = np.linspace(0.001,4,1000)
```

```python
class rainfall(Suite):
    def Likelihood(self,data,hypo):
        x = hypo
        k,t = data
        like = 1/((scipy.special.gamma(k))*t**k)* x**(k-1)* np.exp(-x/t)
        return like
```

```python
data = (.83,.85)
prediction = rainfall(prior)
prediction.Update(data)
thinkplot.Pdf(prediction)
print(prediction.Mean())
```
